```python
# user-friend graph
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

#plt.rcParams['figure.figsize'] = [50,50]

df = pd.read_csv('lastfm-2k/user_friends.csv', sep='\t', na_filter=False)

df
```

```python
values = []

for i in df.index:
    #print(i, j)
    if int(df.loc[i][0]) in user_cluster.index and int(df.loc[i][1]) in user_cluster.index:
        values.append(df.loc[i])
    #if int(df.loc[i][1]) in matrix.index:


len(values)        
```

```python
friend_df = pd.DataFrame(values)

friend_df
```

```python
G = nx.from_pandas_edgelist(friend_df, 'userID','friendID')  
#nx.draw(G, with_labels=True)
#plt.show()
```

```python
# closeness centrality
close_centrality = nx.closeness_centrality(G)
print(len(close_centrality))

close_centrality = pd.DataFrame.from_dict(close_centrality, orient='index', columns=['closeness'])
#close_centrality = close_centrality.sort_values(by=['closeness'], ascending=False) # closeness 높은 순으로 정렬
print(close_centrality)

close_centrality.head(10) # closeness 높은 10명
```

```python
close_centrality.describe()
```

```python
import numpy as np
```

```python
close_centrality_drop = close_centrality[close_centrality.closeness>0.20]
close_centrality_drop
```

```python
close_centrality_drop.max()
close_centrality_drop
```

```python
user_cluster
```

```python
# hyper param

for i in matrix.index:
    if i in close_centrality_drop.index:
            matrix.loc[i][0] = matrix.loc[i][0] * (close_centrality_drop["closeness"][i]/np.exp((-1.5)*close_centrality_drop["closeness"][i]))
            matrix.loc[i][1] = matrix.loc[i][1] * (close_centrality_drop["closeness"][i]/np.exp((-1.5)*close_centrality_drop["closeness"][i]))
```

```python
matrix
```

```python
#KMeans(Viusalization)
km = KMeans(n_clusters=10, init='k-means++')
kcluster = km.fit(matrix)
kcluster_id = pd.DataFrame(kcluster.labels_)

d1 = pd.concat([matrix, kcluster_id], axis=1)
d1.columns = [0, 1, "cluster"]

plt.figure(figsize = (10, 10))
sns.scatterplot(d1[0], d1[1], hue = d1['cluster'], legend="full", palette="deep", marker=".")
sns.scatterplot(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], label = 'Centroids', color = 'red')
plt.title("KMeans Clustering")
plt.legend()
plt.show()


#print('KMeans Silhouette Coefficient: {:.4f}'.format(metrics.silhouette_score(d1.iloc[:,:-1], d1['cluster'])))
#print('KMeans Davies Bouldin Index: {:.4f}'.format(metrics.davies_bouldin_score(d1.iloc[:,:-1], d1['cluster'])))
```
